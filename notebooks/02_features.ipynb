# Data cleaning and preprocessing

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.status.value_counts()
```

```python
LOW_MISSING_SAFE_TO_DROP = ['submission_of_application', 'age', 'loan_purpose', 'neg_ammortization', 'term']
df = df.dropna(subset=LOW_MISSING_SAFE_TO_DROP)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.status.value_counts()
```

```python
LEAKAGE_COLUMNS_DROP = ['upfront_charges', 'interest_rate_spread', 'rate_of_interest']
df = df.drop(columns=LEAKAGE_COLUMNS_DROP)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.status.value_counts()
```

```python
df['income'] = df['income'].fillna(df['income'].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna(subset=['loan_limit', 'approv_in_adv'])
```

```python
df.status.value_counts()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
columns = df.columns
cat_columns = df.columns[df[columns].nunique() < 25]
num_columns = df.columns[df[columns].nunique() >= 25]
df[cat_columns].dtypes
```

```python
df[num_columns].dtypes
```

```python
from sklearn.metrics import mutual_info_score

def mutual_info_default_score(series):
    return mutual_info_score(series, df.status)


mi = df[cat_columns].apply(mutual_info_default_score)
mi = mi.sort_values(ascending=False)
print(mi)
```
